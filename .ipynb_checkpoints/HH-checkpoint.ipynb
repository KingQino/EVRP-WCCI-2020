```python
def elapsed_time(start):
    return time.process_time - start

# Luby sequence generator
def luby(j):
    k = math.ceil(math.log2(j+1))
    l = pow(2, k-1)
    if pow(2, k) == (j+1):
        return l
    else:
        return luby(j-l+1)
    
def generateLuby(maxValue):
    j = 1
    lubyNumber = []
    while 1:
        ind = luby(j)
        lubyNumber.append(ind)
        if ind >= maxValue:
            break
        else:
            j+=1
    return lubyNumber
```
